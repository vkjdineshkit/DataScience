Ex 6 : Write a program to calculate the BMI rate?Get the user input & Result should be in integer. 


```r
weight<-readline(prompt = "Enter the weight=")
weight<-as.integer(weight)
height<-readline(prompt = "Enter the height in cm=")
height<-as.integer(height)
BMI=weight/((height/100)^2)
print(as.integer(BMI))
```

Ex6 : Create a function to calculate the BMI Rating? Result should be in integer?

```r
BMI.function=function(w,h){
  result=w/((h/100)^2)
  result
  print(round(result))
}
w<-readline(prompt = "Enter the weight=")
w<-as.integer(w)
h<-readline(prompt = "Enter the height in cm=")
h<-as.integer(h)
BMI.function(w,h)

```
